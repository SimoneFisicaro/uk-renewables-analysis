```python
import pandas as pd
from time import strptime
```

```python
cons_m_input = pd.read_csv("../data/monthly_quarterly_electricity_consumption_monthly.txt", delimiter="\t", decimal=',')
cons_q_input = pd.read_csv("../data/monthly_quarterly_electricity_consumption_quarterly.txt", delimiter="\t", decimal=',')
```

```python
cons_m_input.head()
```

```python
cons_q_input.head()
```

# Convert year, month, quarter

```python
cons_m_converted = cons_m_input
#cons_m_converted['month'] = cons_m_converted['month'].map(lambda x: strptime(x.replace('*', '').strip(), '%B').tm_mon)

cons_q_converted = cons_q_input
cons_q_converted = cons_q_converted.rename(index=str, columns={"month": "quarter"})
cons_q_converted['quarter'] = cons_q_converted['quarter'].map(lambda x: x.replace('Quarter', '').strip())

cons_q_converted.head()
```

# Filter data

```python
cons_m_filtered = cons_m_converted[cons_m_converted.year >= 2010].reset_index()
cons_q_filtered = cons_q_converted[cons_q_converted.year >= 2010].reset_index()

cons_m_filtered.head()
```

# Save dataframes

```python
cons_m_filtered.to_csv("../data/consumption_monthly.csv")
cons_q_filtered.to_csv("../data/consumption_quarterly.csv")
```
